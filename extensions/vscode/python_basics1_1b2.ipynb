### 5. ë°˜ë³µë¬¸ ì‹¤ìŠµ
ì‚¬ìš©ìê°€ ì…ë ¥í•œ ë²”ìœ„ì˜ ìˆ«ìë“¤ì— ëŒ€í•´ ë‹¤ìŒì„ ê³„ì‚°í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì„¸ìš”.

ìš”êµ¬ì‚¬í•­:
- ì‹œì‘ ìˆ«ìì™€ ë ìˆ«ìë¥¼ ì…ë ¥ë°›ìŒ
- í•´ë‹¹ ë²”ìœ„ì˜ ëª¨ë“  ìˆ«ìì˜ í•©ê³„
- ì§ìˆ˜ì˜ í•©ê³„
- í™€ìˆ˜ì˜ í•©ê³„
- í‰ê· ê°’
- ê³„ì‚° ê³¼ì •ì„ ì‹œê°ì ìœ¼ë¡œ í‘œì‹œ

````python
# ì‹¤ìŠµ 5 í’€ì´
print("=== ìˆ«ì ë²”ìœ„ ê³„ì‚° í”„ë¡œê·¸ë¨ ===\n")

# ë²”ìœ„ ì…ë ¥
start = int(input("ì‹œì‘ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
end = int(input("ë ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))

# ë³€ìˆ˜ ì´ˆê¸°í™”
total = 0
even_sum = 0
odd_sum = 0
even_count = 0
odd_count = 0

# ê³„ì‚° ê³¼ì • ì¶œë ¥
print("\nê³„ì‚° ê³¼ì •:")
for i in range(start, end + 1):
    total += i
    if i % 2 == 0:
        even_sum += i
        even_count += 1
    else:
        odd_sum += i
        odd_count += 1
    
    if i < end:
        print(i, end=" + ")
    else:
        print(i)

# ê²°ê³¼ ì¶œë ¥
count = end - start + 1
print("\nê³„ì‚° ê²°ê³¼:")
print(f"- ì „ì²´ í•©ê³„: {total}")
print(f"- ì§ìˆ˜ í•©ê³„: {even_sum}")
print(f"- í™€ìˆ˜ í•©ê³„: {odd_sum}")
print(f"- í‰ê· ê°’: {total/count}")

print("\nì¶”ê°€ ì •ë³´:")
print(f"- ì§ìˆ˜ ê°œìˆ˜: {even_count}ê°œ")
print(f"- í™€ìˆ˜ ê°œìˆ˜: {odd_count}ê°œ", end="")
````

### 6. ë°ì´í„° êµ¬ì¡° ì‹¤ìŠµ
í•™ìƒë“¤ì˜ ì„±ì  ì •ë³´ë¥¼ ê´€ë¦¬í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ì‘ì„±í•˜ì„¸ìš”.

ìš”êµ¬ì‚¬í•­:
- í•™ìƒ ì´ë¦„ê³¼ ì ìˆ˜ë¥¼ ì…ë ¥ë°›ì•„ ë”•ì…”ë„ˆë¦¬ì— ì €ì¥
- ì „ì²´ í•™ìƒ ëª©ë¡ê³¼ ì ìˆ˜ ì¶œë ¥
- ìµœê³ ì ê³¼ ìµœì €ì  ì¶œë ¥
- í‰ê·  ì ìˆ˜ ê³„ì‚°
- íŠ¹ì • ì ìˆ˜ ì´ìƒì¸ í•™ìƒë“¤ì˜ ëª…ë‹¨ ì¶œë ¥

````python
# ì‹¤ìŠµ 6 í’€ì´
print("=== í•™ìƒ ì„±ì  ê´€ë¦¬ í”„ë¡œê·¸ë¨ ===\n")

# í•™ìƒ ì •ë³´ ì €ì¥í•  ë”•ì…”ë„ˆë¦¬
students = {}

# í•™ìƒ ì •ë³´ ì…ë ¥
while True:
    name = input("í•™ìƒ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” (ì¢…ë£Œí•˜ë ¤ë©´ q): ")
    if name.lower() == 'q':
        break
    score = int(input(f"{name}ì˜ ì ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))
    students[name] = score

# ì „ì²´ í•™ìƒ ì„±ì  ì¶œë ¥
print("\nì „ì²´ í•™ìƒ ì„±ì :")
for name, score in students.items():
    print(f"- {name}: {score}ì ")

# ì„±ì  ë¶„ì„
max_score = max(students.values())
min_score = min(students.values())
avg_score = sum(students.values()) / len(students)

# ìµœê³ ì ê³¼ ìµœì €ì  í•™ìƒ ì°¾ê¸°
max_student = [name for name, score in students.items() if score == max_score][0]
min_student = [name for name, score in students.items() if score == min_score][0]

print("\nì„±ì  ë¶„ì„:")
print(f"- ìµœê³ ì : {max_score}ì  ({max_student})")
print(f"- ìµœì €ì : {min_score}ì  ({min_student})")
print(f"- í‰ê· ì ìˆ˜: {avg_score}ì ")

# 80ì  ì´ìƒ í•™ìƒ ì¶œë ¥
print("\n80ì  ì´ìƒ í•™ìƒ ëª…ë‹¨:")
high_scores = {name: score for name, score in students.items() if score >= 80}
for name, score in high_scores.items():
    print(f"- {name}: {score}ì ", end="")
````
